```python
from PIL import Image
import numpy as np
import sys
import os
import csv
import pandas as pd

def createFileList(myDir, format='.jpg'):
    fileList = []
    print(myDir)
    for root, dirs, files in os.walk(myDir, topdown=False):
        for name in files:
            if name.endswith(format):
                fullName = os.path.join(root, name)
                fileList.append(fullName)
    return fileList

myFileList = createFileList('/Users/김동휘/Desktop/머신러닝/term project/data/train/melanoma/')

for file in myFileList:
    print(file)
    img_file = Image.open(file)

    width, height = 28, 28
    format = img_file.format
    mode = img_file.mode

    img_grey = img_file.convert('L')
    img_grey = img_grey.resize((28, 28))

    value = np.asarray(img_grey.getdata(), dtype=np.int).reshape((img_grey.size[1], img_grey.size[0]))
    value = value.flatten()

    with open("melanoma.csv", 'a') as f:
        writer = csv.writer(f)
        writer.writerow(value)
        
df = pd.read_csv('/Users/김동휘/Desktop/머신러닝/term project/melanoma.csv')
df.head()
```
